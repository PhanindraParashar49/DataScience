```python
def get_data(scheme_code,mf=mf):
    df = mf.get_scheme_historical_nav(scheme_code,as_Dataframe=True).reset_index()
    df.rename(columns={'nav':scheme_code,'dayChange':f'{scheme_code}_change'},inplace=True)
    return df

def get_nav_list(scheme_codes,mf=mf):
    data = mf.get_scheme_historical_nav(scheme_codes[0],as_Dataframe=True).reset_index().rename(columns={'nav':scheme_codes[0]}).drop(['dayChange'],axis=1)

    for i in scheme_codes[1:]:
        data_ = mf.get_scheme_historical_nav(i,as_Dataframe=True).reset_index().rename(columns={'nav':i}).drop(['dayChange'],axis=1)
        data = pd.merge(data,data_,on='date',how='inner')
    data.index = list(range(len(data)))
    data['date'] = pd.to_datetime(data['date'])
    return data
```

```python
dn = get_nav_list(scheme_codes=list(all_scheme_codes.keys())[:10])
```
